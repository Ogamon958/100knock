```python
#20
import json 

file_name = 'jawiki-country.json'
for line in open(file_name, mode='r'):
    line = json.loads(line) #json形式の各行を辞書形式に変換
    if line['title'] == 'イギリス': #辞書で検索
      text_uk = line['text']
      break
```

```python
#20(確認)
print(text_uk) #本文を出力
```

```python
#21
import re

pattern = r'^(.*?\[\[Category:.*?\]\].*)$' # https://qiita.com/luohao0404/items/7135b2b96f9b0b196bf3 正規表現の参照リスト []を正規表現で用いる際エスケープが必要なことに注意 https://qiita.com/katsukii/items/1c1550f064b4686c04d4
result = '\n'.join(re.findall(pattern, text_uk, re.MULTILINE)) #re.MULTILINEを加えると各行の先頭にマッチするようになる、デフォルトだと文字列全体の先頭のみ
print(result)
```

```python
#22
import re

pattern = r'^.*?\[\[Category:(.*?)(?:\|.*)?\]\].*$' #(?:\|.*)?の部分、?:を初めにつけると、その部分はマッチはするが出力はしない、|以下が邪魔なのでその箇所を除去するために追加、|がないこともあるため、後ろの？も必要
result = '\n'.join(re.findall(pattern, text_uk, re.MULTILINE))
print(result)
```

```python
#23 (== のときレベル1 ===の時レベル2...)
pattern = r'^(\={2,})\s*(.+?)\s*(\={2,}).*$' #^(\={2,})は=が2個以上の意、\s*は空白が存在することもあるため追加 https://murashun.jp/article/programming/regular-expression.html
result = '\n'.join(i[1] + ':' + str(len(i[0]) - 1) for i in re.findall(pattern, text_uk, re.MULTILINE))#i[0]→(\={2,})、i[1]→(.+?)、i[2]→(\={2,})、 =の数は左右でそろっているので、(len(i[2]) - 1)としても同様の結果となる。
print(result)
```

```python
#24
pattern = r'\[\[ファイル:(.+?)\|'
result = '\n'.join(re.findall(pattern, text_uk)) #文頭以外にもファイルあるので、MULTILINEは適応しない
print(result)
```

```python
#25
# テンプレートの抽出
pattern = r'^\{\{基礎情報 国(.*?)^\}\}'  #文章の初めに {{基礎情報 国 が、次に現れるある文章の初めに }} が付いているものが基礎情報の最後
template = re.findall(pattern, text_uk, re.MULTILINE + re.DOTALL) #re.DOTALLについて https://qiita.com/FukuharaYohei/items/459f27f0d7bbba551af7 (基礎情報の中に改行が含まれるため)
print(template)

# フィールド名と値を辞書オブジェクトに格納
pattern = r'^\|(.+?)\s*=\s*(.+?)(?:(?=\n\|)|(?=\n$))' #\n|か、最後の場合の\n$は出力しないようにする。その他、空白部分も省略する。
result = dict(re.findall(pattern, template[0], re.MULTILINE + re.DOTALL)) # 上記のtemplateに対し、正規表現を含む部分を抜き出し、辞書化する。0番目の()内がkey,1番目の()内がvalueになる。#re.findall+dict https://atmarkit.itmedia.co.jp/ait/articles/2103/05/news028.html #re.DOTALLについて https://qiita.com/FukuharaYohei/items/459f27f0d7bbba551af7
for k, v in result.items(): #辞書のforループ処理 https://note.nkmk.me/python-dict-keys-values-items/
  print(k + ': ' + v)
```
