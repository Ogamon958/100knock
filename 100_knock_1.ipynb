```python
#4(辞書型)
import re
str = 'Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.'
re_str = re.sub('[,.]', '', str)
split_str = re.split(' ', re_str) 
print(split_str)
one_ch_list = [1, 5, 6, 7, 8, 9, 15, 16, 19]  # 1文字を取り出す単語の番号リスト
ans = {}
for i, word in enumerate(split_str):
  if i + 1 in one_ch_list:
    ans[word[:1]] = i + 1  # リストにあれば1文字を取得
  else:
    ans[word[:2]] = i + 1  # なければ2文字を取得

print(ans)
```

```python
#4(マップ型)
import re
str = 'Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.'
re_str = re.sub('[,.]', '', str)
split_str = re.split(' ', re_str) 
print(split_str)
one_ch = [1, 5, 6, 7, 8, 9, 15, 16, 19]  # 1文字を取り出す単語の番号リスト
ans = list(map(lambda word, i: word[:1] if i+1 in one_ch else word[:2], split_str, range(len(split_str))))

print(ans)
```

```python
#5
def ngram(n, lst):
  # ex.
  # [str[i:] for i in range(2)] -> ['I am an NLPer', ' am an NLPer']
  # zip(*[str[i:] for i in range(2)]) -> zip('I am an NLPer', ' am an NLPer')
  return list(zip(*[lst[i:] for i in range(n)]))

str = 'I am an NLPer'
words_bi_gram = ngram(2, str.split()) #空白文字で分割
chars_bi_gram = ngram(2, str)

print('単語bi-gram:', words_bi_gram)
print('文字bi-gram:', chars_bi_gram)
```

```python
#6
def ngram(n, lst):
  return list(zip(*[lst[i:] for i in range(n)]))

str1 = 'paraparaparadise'
str2 = 'paragraph'

chars_bi_gram1 = set(ngram(2, str1))
chars_bi_gram2 = set(ngram(2, str2))

union = chars_bi_gram1 | chars_bi_gram2
intersection = chars_bi_gram1 & chars_bi_gram2
difference = chars_bi_gram1 - chars_bi_gram2


print('X:', chars_bi_gram1)
print('Y:', chars_bi_gram2)
print('和集合(X | Y):', union)
print('積集合(X & Y):', intersection)
print('差集合(X - Y):', difference)


```

```python
#8
def cipher(str):
  cryptography = [chr(219 - ord(x)) if x.islower() else x for x in str]

  return ''.join(cryptography)

sentence = 'the quick brown fox jumps over the lazy dog'
sentence = cipher(sentence)
print('暗号化:', sentence)
sentence = cipher(sentence)
print('復号化:', sentence)
```

```python
#9
import random
def word_shuffle(words):
  result = []
  for word in words.split():
    if len(word) > 4:
      word = word[:1] + ''.join(random.sample(word[1:-1], len(word) - 2)) + word[-1:] #並び変える
    result.append(word) #リストに追加

  return ' '.join(result) #結合

words = "I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."
ans = word_shuffle(words)

print(ans)
```
